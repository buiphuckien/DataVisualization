# How to plot a bar chart with python

The table below show the age of addicts at an addicted center on years 1971 and 1972.

|        ages      |1971  |1972  |
|------------------|------|------|
|   under 20       |118   |96    |
|   20 - under 25  |772   |727   |
|   25 - under 30  |288   |376   |
|   30 - under 35  |112   |117   |
|   35 - under 50  |112   |118   |
|   above 50       |117   |165   |
<ol type="a">
  <li>Plot a bar chart for addiction of the year 1971.</li>
  <li>Plot a bar chart to compare the addiction of the tow years</li>
<ol>

```python
# Import the libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

# Create arrays for age, number of addicts in 1971, and number of addicts in 1972
age = ["under 20", "20 - under 25", "25 - under 30", "30 - under 35", "35 - under 50", "above 50"]
num_of_people_1971 = [118, 772, 288, 112, 112, 117]
num_of_people_1972 = [96, 727, 376, 117, 118, 165]

# The indices for the bars
ind = np.arange(len(age))

# The with of the bars
width = 0.35

# Create the plots
fig, ax = plt.subplots(figsize=(12, 6))

# Create a horizontal bar chart
ax.barh(age, num_of_people_1972)

# Set the label for x-axis
ax.set_xlabel('Number of people', fontweight = 'bold')

# Set the label for y-axis
ax.set_ylabel('Ages', fontweight = 'bold')

# Set y-ticks and y-ticklabels
ax.set_yticks(ind)
ax.set_yticklabels(age)

# Add labels for patches
def add_value_labels(ax, spacing=5):  
    # each bar in patches
    for rect in ax.patches:
        # Get x_value, y_value for the labels
        x_value = rect.get_width()
        y_value = rect.get_y() + rect.get_height()/4

        # The distance between bars and lables
        space = spacing

        # Use x_value as the label
        label = "{:}".format(x_value)

        # Create annotate
        ax.annotate(
            label,                      # label
            (x_value, y_value),         # Position of the label
            xytext=(space, 0),          # Distance between label and bar
            textcoords="offset points") # Interpret `xytext` as offset in points

# Call the function above.
add_value_labels(ax)

ax.set_title(
    'Number of addicts at the center in year 1972', 
    fontsize = 15,
    fontweight = 'bold',
    pad = 20,
    loc = 'center')
plt.show()
```

```python
# Create subplots
fig, ax = plt.subplots(figsize=(12, 6))
width = 0.35

# Plot horizontal bar chart for both years
plot1 = ax.barh(ind - width/2, num_of_people_1971, width, label='Year 1971')
plot2 = ax.barh(ind + width/2, num_of_people_1972, width, label='year 1972')

# Call add_value_lables function
add_value_labels(ax)
ax.set_xlabel('Number of people', fontweight = 'bold')
ax.set_ylabel('Ages', fontweight = 'bold')
ax.set_yticks(ind)
ax.set_yticklabels(age)
ax.set_title(
    'The age of addicts at an addicted center on years 1971 and 1972',
    fontsize = 15,
    fontweight = 'bold',
    pad = 20,
    loc = 'center')
plt.legend()
plt.show()
```

The graph potray the ages of addicts at an addicted center in years 1971 and 1972. <br>
At this time, most of addicts were in the age of 20 to 25 - account for more than 50% of total number of people here. Additionally, about 60% of people here were less than 25. This shows addiction was a serious problem for young people. <br>
From year 1971 to 1972, number of addicts decreased in the group of ages under 25. It vice versed for other groups. Especially for the group of ages that above 50, number of addicts rapidly increased.
