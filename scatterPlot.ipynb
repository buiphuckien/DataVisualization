```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

# Create two arrays - one for year and the other for revenue
year = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
revenue = [42905, 65225, 108249, 156508, 170910, 182795, 233715, 215639, 229234, 265595, 260174, 274515, 365817]

# Create a plot
fig, ax = plt.subplots(figsize=(12, 6))

# Create a scatter plot using triangular markers
ax.scatter(year, revenue, marker='^')

# Set label for x-axis
ax.set_xlabel('Year', fontweight = 'bold')

# Set label for y-axis
ax.set_ylabel('Revenue', fontweight = 'bold')

# Set title for the plot
ax.set_title(
    'Apple Annual Revenue (Millions of US $)', 
    fontsize = 15,
    fontweight = 'bold',
    pad = 20,
    loc = 'center')
plt.show()
```

The chart illustrates yearly revenue of Apple from the year of 2009 to 2021. </br>

Through the chart we see Apple annual revenue increased steadily from 2009 to 2015. However, it slightly declined in the year of 2016. The revenue climbed slowly for the following years. It shot up in the year of 2021 with more than 35 billions and marked the highest annualrevenue in history of this company.
