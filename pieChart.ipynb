# How to plot a pie chart with python

The table below show the age of addicts at an addicted center on years 1971 and 1972.

|        ages      |1971  |1972  |
|------------------|------|------|
|   under 20       |118   |96    |
|   20 - under 25  |772   |727   |
|   25 - under 30  |288   |376   |
|   30 - under 35  |112   |117   |
|   35 - under 50  |112   |118   |
|   above 50       |117   |165   |

Plot pie charts the year of 1971.

```python
# Import the libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

# Create arrays for ages and number of addicts in 1971
age = ["under 20", "20 - under 25", "25 - under 30", "30 - under 35", "35 - under 50", "above 50"]
num_of_people_1971 = [118, 772, 288, 112, 112, 117]

# Create a plot
fig, ax = plt.subplots(figsize=(12, 6))

def func(pct, num_of_people):
    absolute = int(np.round(pct/100.*np.sum(num_of_people_1971)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

wedges, texts, autotexts = ax.pie(num_of_people_1971, autopct=lambda pct: func(pct, num_of_people_1971),
                                  textprops=dict(color="w"))

ax.legend(wedges, age,
          title="Ages",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=10)

ax.set_title(
    "Number of addicts at the center in year 1972",
    fontsize=15,
    fontweight = 'bold',
    loc = 'center')
plt.show()
```
